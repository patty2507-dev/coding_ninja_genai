```python
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.openai import OpenAI
Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    
)

Settings.llm = OpenAI(model="gpt-4o-mini")
from llama_index.core import SimpleDirectoryReader
from llama_index.core import VectorStoreIndex
from llama_index.core.response.pprint_utils import pprint_response


```

```python

documents = SimpleDirectoryReader('data_test').load_data()
index = VectorStoreIndex(documents)
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is a personal loan?")
pprint_response(response,show_source=True)
# response
```

```python
from llama_index.core.node_parser import SentenceSplitter
splitter = SentenceSplitter(
    chunk_size=512,
    chunk_overlap=100
)

nodes = splitter.get_nodes_from_documents(documents)
print(f"  Nodes created : {len(nodes)}")
```

```python
# indexing 
# VectorStoreIndex : semnatic qnA
# SummaryIndex : Full doc overview
# KeywordIndex : Exact facts
# TreeIndex : tree structure
```

```python
from llama_index.core import (
    VectorStoreIndex, SummaryIndex,
    KeywordTableIndex, TreeIndex
)
```

```python
vector_index = VectorStoreIndex(nodes)
summar_index = SummaryIndex(nodes)
keyword_index = KeywordTableIndex(nodes)
tree_index = TreeIndex(nodes)
```

```python
rv= vector_index.as_query_engine()
sv = summar_index.as_query_engine()
kv = keyword_index.as_query_engine()
tv = tree_index.as_query_engine()
```

```python
q = "How many days is maternity leave?"

print("vector index")
pprint_response(rv.query(q),show_source=True)

print("keyword index")
pprint_response(kv.query(q),show_source=True)
print("tree index")
pprint_response(tv.query(q),show_source=True)
```

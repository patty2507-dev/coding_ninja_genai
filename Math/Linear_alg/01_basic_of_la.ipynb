```python
def linear_score(f1,f2,w1,w2,w0):
    return w0 + w1*f1 + w2*f2


def predict(f1,f2,w0,w1,w2):
    output = linear_score(f1,f2,w0,w1,w2)
    if output >0:
        return 1
    else:
        return 0
```

```python
import numpy as np
not_spam = np.array([
    [1, 1],
    [1, 2],
    [2, 1],
    [2, 2],
    [3, 1],
])

spam = np.array([
    [6, 6],
    [7, 5],
    [6, 7],
    [8, 6],
    [7, 7],
])
```

```python
# Manual weights and bias
w1, w2, w0 = 1.0, 1.0, 0
```

```python
for i in range(len(spam)):
    f1 = spam[i][0]
    f2 = spam[i][1]
    x = predict(f1,f2,w0,w1,w2)
    print(x)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(spam[:, 0], spam[:, 1], color='blue')
ax.scatter(not_spam[:, 0], not_spam[:, 1], color='red')
# plt.savefig('01_basic_of_la.png', dpi=150, bbox_inches='tight')
x_values = np.linspace(0, 10, 100)
y_vals= -(w0 + w1*x_values)/w2
ax.plot(x_values, y_vals, color='black')
# ewuqtion of line
# y=mx+c
```

```python
from gensim.models import Word2Vec
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
import gensim.downloader as api
wiki_model = api.load("glove-wiki-gigaword-50")
```

```python
vector = wiki_model["loan"]
```

```python
len(vector)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)


words = ["loan", "credit", "finance", "money", "gold", "bank", "banana"]

word_vector = [wiki_model[word] for word in words]
points_2d = pca.fit_transform(word_vector)
```

```python
points_2d
```

```python
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    x = points_2d[i, 0]
    y = points_2d[i, 1]
    plt.scatter(x, y)
    plt.text(x + 0.01, y + 0.01, word, fontsize=10)

plt.title("PCA of Pretrained Word Embeddings")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```

```python
import yaml

with open('agents.yaml','r') as f:
    agent_config = yaml.safe_load(f)

with open('tasks.yaml','r') as f:
    task_config = yaml.safe_load(f)
```

```python
agent_config
```

```python
agent_config['customer_support_agent']['goal'],
```

```python
customer_support_agent = Agent(
    role=agent_config['customer_support_agent']['role'],
    goal=agent_config['customer_support_agent']['goal'],
    backstory=agent_config['customer_support_agent']['backstory'],
    tools=[],
    llm = "gpt-4o-mini",
    verbose=True
)

support_task = Task(
    description=task_config["support_task"]["description"],
    expected_output=task_config["support_task"]["expected_output"],
    agent=customer_support_agent
)
```

```python
crew = Crew(
    agents=[customer_support_agent],
    tasks=[support_task],
    process=Process.sequential
)

result = await crew.kickoff_async({"customer_query": 'I bought an item and I dont like it. I want to return it. What is your return policy for this case?'})

print(result.raw)
```

```python
from langchain_community.document_loaders import PyPDFLoader,CSVLoader
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter
)
```

```python
loader = PyPDFLoader("bajaj_finance_policy_prose_v1.pdf")
# csv_loader = CSVLoader("sample_docs/customers.csv")
docs = loader.load()
# csv_docs = csv_loader.load()
```

```python
# print(docs[1].page_content)
```

```python
s1 = CharacterTextSplitter(
    separator=".",
    chunk_size=200, 
    chunk_overlap=0)

chunks = s1.split_documents(docs)
```

```python
s2 = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ".", ","],
    chunk_size=500, 
    chunk_overlap=0)

chunks = s2.split_documents(docs)
    
```

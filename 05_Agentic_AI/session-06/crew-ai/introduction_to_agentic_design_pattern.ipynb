```python
import yaml

with open('agents.yaml','r') as f:
    agent_config = yaml.safe_load(f)

with open('tasks.yaml','r') as f:
    task_config = yaml.safe_load(f)
```

```python
agent_config
```

```python
agent_config['customer_support_agent']['goal'],
```

```python
customer_support_agent = Agent(
    role=agent_config['customer_support_agent']['role'],
    goal=agent_config['customer_support_agent']['goal'],
    backstory=agent_config['customer_support_agent']['backstory'],
    tools=[],
    llm = "gpt-4o-mini",
    verbose=True
)

support_task = Task(
    description=task_config["support_task"]["description"],
    expected_output=task_config["support_task"]["expected_output"],
    agent=customer_support_agent
)
```

```python
crew = Crew(
    agents=[customer_support_agent],
    tasks=[support_task],
    process=Process.sequential
)

result = await crew.kickoff_async({"customer_query": 'I bought an item and I dont like it. I want to return it. What is your return policy for this case?'})

print(result.raw)
```

```python
with open('multiple_tasks.yaml','r') as f:
    task_config = yaml.safe_load(f)
```

```python
agent_details = agent_config["customer_support_agent"]
customer_support_agent = Agent(
    role=agent_details["role"],
    goal=agent_details["goal"],
    backstory=agent_details["backstory"],
    tools=[],
    verbose=True,
)


return_task_config = task_config["return_task"]
return_task = Task(
    description=return_task_config["description"],
    expected_output=return_task_config["expected_output"],
    agent=customer_support_agent
)

shipping_task_config = task_config["shipping_task"]
shipping_task = Task(
    description=shipping_task_config["description"],
    expected_output=shipping_task_config["expected_output"],
    agent=customer_support_agent
)


complaint_task_config = task_config["complaint_task"]
complaint_task = Task(
    description=complaint_task_config["description"],
    expected_output=complaint_task_config["expected_output"],
    agent=customer_support_agent
    )

email_task_config = task_config["email_task"]
email_task = Task(
    description=email_task_config["description"],
    expected_output=email_task_config["expected_output"],
    agent=customer_support_agent
    )
```

```python
crew = Crew(
    agents=[customer_support_agent],
    tasks=[return_task, shipping_task, complaint_task, email_task],
    process=Process.sequential,
    memory=True
)
```

```python
results = await crew.kickoff_async({"customer_inquiry": "I bought laptop but this is not working, can you help me"})
```

```python
print(results.raw)
```

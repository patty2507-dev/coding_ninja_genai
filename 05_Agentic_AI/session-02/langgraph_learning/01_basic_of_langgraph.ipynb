```python
from typing import Annotated, TypedDict
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage
from langgraph.graph.message import add_messages

class AgentState(TypedDict):
    messages: Annotated[List[str], append_reducer]
    verdict: str


current_messages = [HumanMessage(content="Review PR #1")]
ai_response = AIMessage(content="", tool_calls=[
    {"name": "get_pr_details", "args": {}, "id": "call_1", "type": "tool_call"}
])

tool_result = ToolMessage(
    content='{"title": "Fix SQL injection", "author": "rahul8879"}',
    tool_call_id="call_1",
    name="get_pr_details"
)



current_messages = add_messages(current_messages, [tool_result])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('customer_data.csv')
data.head(2)
data.info()
```

```python
def clean_income(x):
    return int(x.replace("₹",'').replace(',',''))

data['Monthly Income (₹)'] = data['Monthly Income (₹)'].apply(clean_income)
```

```python
X = data[['CIBIL Score', 'Monthly Income (₹)']]
target = data['Loan Status']

x_train, x_test,y_train,y_test = train_test_split(X,target,test_size=0.10)
```
